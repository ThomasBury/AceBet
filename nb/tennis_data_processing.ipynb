```python
def permute_data(data):
    permuted_data = []
    for index, row in data.iterrows():
        r = random.randint(0, 1)
        if r == 0:
            rta = row.tolist()
            rta.append(0)
        else:
            rta = row.tolist()
            rta[3], rta[4] = rta[4], rta[3]
            rta[5], rta[6] = rta[6], rta[5]
            rta[7], rta[8] = rta[8], rta[7]
            rta[9], rta[10] = rta[10], rta[9]
            rta[11], rta[12] = rta[12], rta[11]
            rta.append(1)
        permuted_data.append(rta)
    return permuted_data
```

```python
selected_columns = [
    "Date",
    "Surface",
    "Best of",
    "Player0",
    "Player1",
    "0Rank",
    "1Rank",
    "PS0",
    "PS1",
    "B3650",
    "B3651",
    "elo_0",
    "elo_1",
    "winner",
]
permuted_df = pd.DataFrame(columns=selected_columns)
```

```python
df = pd.DataFrame(permute_data(data_no_nan), columns=selected_columns)
```
