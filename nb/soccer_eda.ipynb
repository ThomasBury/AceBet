# Database schema

```python
import pandas as pd
import sqlite3

path = "../data/"  # Insert path here
database = path + "soccer.sqlite"

conn = sqlite3.connect(database)

query = """SELECT *
        FROM sqlite_master
        WHERE type='table';"""

tables = pd.read_sql(query, conn)
tables
```

# EDA

## Permier League

```python
query = """SELECT *
        FROM Country;"""

countries = pd.read_sql(query, conn)
print(f"Is the England part of the DB? {'England' in countries.name.unique()}")

countries
```

```python
query = """SELECT country_id, League.name as league_name, Country.name as country_name
        FROM League
        JOIN Country ON Country.id = League.country_id;"""

leagues = pd.read_sql(query, conn)
print(
    f"Is the England part of the DB? {'England Premier League' in leagues.league_name.unique()}"
)

leagues
```

## Arsenal

```python
query = """SELECT *
        FROM Team
        WHERE team_long_name LIKE 'Arsenal' ;"""
teams = pd.read_sql(query, conn)
teams
```
